```python
user_data = data.pivot_table(index = 'User Id',
                       columns = 'Book.Title',
                       values = 'Book.Rating').reset_index(drop=True)
```

```python
user_data
```

```python
user_data.index = data['User Id'].unique()
```

```python
user_data
```

```python
user_data.fillna(0, inplace=True)
```

```python
user_data
```

```python
# Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_data.values, metric='cosine')
```

```python
user_sim
```

```python
user_sim_data = pd.DataFrame(user_sim)
```

```python
user_sim_data.head(10)
```

```python
user_sim_data.index = data['User Id'].unique()
user_sim_data.columns = data['User Id'].unique()
```

```python
user_sim_data.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_data.iloc[0:5, 0:5]
```

```python
#Most similar Users
user_sim_data.idxmax(axis=1)[0:10]
```
